```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_tracks=pd.read_csv("tracks.csv")
df_tracks.head()
```

```python
#Check for null values
pd.isnull(df_tracks).sum()
```

```python
df_tracks.describe().transpose()
```

# Q1- 10 least popular songs

```python
df_sorted=df_tracks.sort_values("popularity",ascending=True).head(10)
df_sorted
```

# Q2- Most popular songs

```python
most_popular=df_tracks.query('popularity>90',inplace=False).sort_values('popularity',ascending=False)
most_popular[:10]
```

```python
#change index to date time
df_tracks=df_tracks.set_index("release_date")
```

```python
df_tracks.index=pd.to_datetime(df_tracks.index,errors="coerce",infer_datetime_format=True)
```

```python
df_tracks
```

```python
df_tracks[["artists"]].iloc[18]
```

# Q3- convert milliseconds to seconds

```python
df_tracks["duration"]=df_tracks["duration_ms"].apply(lambda x: round(x/1000))
df_tracks.drop("duration_ms",inplace=True,axis=1)
```

```python
df_tracks.duration.head()
```

# Q4- correlation map

```python
corr_df=df_tracks.drop(["key","mode","explicit","id_artists","id","artists","name"],axis=1).corr(method="pearson")
plt.figure(figsize=(14,6))
heatmap=sns.heatmap(corr_df,annot=True,fmt=".1g",vmin=-1,vmax=1,center=0,cmap="inferno",linewidths=1,linecolor="Black")
heatmap.set_title("Correlation map between variables")
heatmap.set_xticklabels(heatmap.get_xticklabels(),rotation=90)
```

# Q5- sample resgression line

```python
sample_df=df_tracks.sample(int(0.004*len(df_tracks)))
len(sample_df)
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df,y="loudness",x="energy", color="c").set(title="Loudness Vs Energy Correlation")
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df,y="popularity",x="acousticness", color="r").set(title="Popularity Vs Acousticness Correlation")
```

```python
df_tracks["dates"]=df_tracks.index.get_level_values('release_date')
df_tracks.dates=pd.to_datetime(df_tracks.dates,errors="coerce",infer_datetime_format=True)
years=df_tracks.dates.dt.year
```

```python
df_tracks
```

```python
years
```

```python
sns.displot(years,discrete=True,aspect=2, height=5, kind="hist").set(title="number of songs per year")
```

```python
total_dr=df_tracks.duration
fig_dims=(18,7)
fig,ax=plt.subplots(figsize=fig_dims)
fig=sns.barplot(x=years,y=total_dr,ax=ax,errwidth=False).set(title="Years & Duartion")
plt.xticks(rotation=90)
```

```python
years1=years.copy()
years1=years1.drop_duplicates()
years1
```

```python
total_dr=df_tracks.duration
fig_dims=(10,5)
sns.set_style(style="whitegrid")
fig,ax= plt.subplots(figsize=fig_dims)
fig=sns.lineplot(x=years1,y=total_dr,ax=ax).set(title="Years Vs Duration")
plt.xticks(rotation=60)
```

# Genre Dataset

```python
df_genre=pd.read_csv("SpotifyFeatures.csv")
df_genre
```

```python
plt.title("duration of the songs in different genres")
sns.color_palette("rocket", as_cmap=True)
sns.barplot(y="genre",x="duration_ms",data=df_genre)
plt.xlabel("Duration in ms")
plt.ylabel("Genres")
```

```python
sns.set_style(style="darkgrid")
plt.figure(figsize=(10,5))
famous=df_genre.sort_values('popularity',ascending=False).head(10)
sns.barplot(y="genre",x="popularity",data=famous).set(title="Top 5 Genres by popularity")
```
